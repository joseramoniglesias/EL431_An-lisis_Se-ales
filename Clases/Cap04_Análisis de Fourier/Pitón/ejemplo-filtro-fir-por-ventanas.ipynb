```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
n  = 81
fc = 0.2
b  = signal.firwin(n, fc, window='hann')
print(b)
```

```python
w, h = signal.freqz(b)
fig, ax1 = plt.subplots()
ax1.plot(w/(max(w)), 20 * np.log10(abs(h)), 'b')
ax1.set_title('Respuesta en frecuencia FIR por ventana')
ax1.set_ylabel('Amplitud [dB]', color='b')
ax1.set_xlabel('Frecuencia [f/fs]')
ax2 = ax1.twinx()
angles = np.unwrap(np.angle(h))
ax2.plot(w/(max(w)), angles, 'orange')
ax2.set_ylabel('Fase (radianes)', color='orange')
ax2.grid()
ax2.axis('tight')
plt.show()
```

```python
muestras = 75
escalon = np.ones(muestras)
escalon[0:int(muestras//32)]=0
plt.figure()
plt.stem(escalon, basefmt=' ', use_line_collection=True, label='x[n]')
plt.plot(signal.lfilter(b,1,escalon), 'r-o', label='y[n]')
plt.grid(True)
plt.title('Filtro FIR por ventana')
plt.xlabel('muestras')
plt.ylabel('salida [arbitrario]')
plt.legend()
plt.show()
```
